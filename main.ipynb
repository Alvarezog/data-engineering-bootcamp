```python
def count_categories(series):
    """This function iterates through the chains column, verifies if the name is inside a
    list that we will create as we move through the column, add the name to the list if it isn't already in it
    and add 1 to the counter"""
    #Defining non local variable outside of the scope of this function to make this a nonpure function.
    global count
    chains = []
    for i in series:
        if i in chains:
            pass
        else:
            chains.append(i)
            count += 1
        
    return count
```

```python
#Function that retrun the length of a pandas series
def len_series(series):
    length = 0
    for i in series:
        length += 1
    
    return length
```

```python
def count_values(series):
    states = dict()
    for i in range(len_series(series.iloc[:,0])):
        if series.loc[i]['estado'] in states and series.loc[i]['producto'] in states[series.loc[i]['estado']]:
            states[series.loc[i]['estado']][series.loc[i]['producto']] = states[series.loc[i]['estado']][series.loc[i]['producto']] + 1
        elif series.loc[i]['estado'] in states and series.loc[i]['producto'] not in states[series.loc[i]['estado']]:
            states[series.loc[i]['estado']][series.loc[i]['producto']] = 1
        else:
            states[series.loc[i]['estado']]= {}
            states[series.loc[i]['estado']][series.loc[i]['producto']] = 1
    return states
```

```python
def top_values(dictionary):
    dic_list = []
    for k,v in dictionary.items():
        dic_list.append([k,v])
            
    for i in range(len_series(dic_list)):
     
        max_indx = i
        for j in range(i+1, len_series(dic_list)):
            if  dic_list[j][1] > dic_list[max_indx][1]:
                max_indx = j
                 
        dic_list[i], dic_list[max_indx] = dic_list[max_indx], dic_list[i]
 
    return dic_list
        
```
