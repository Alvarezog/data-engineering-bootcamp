```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/sample.csv')
```

```python
data.head()
```

```python
count = 0
def count_chains(series):
    chains = []
    for i in series:
        if i in chains:
            pass
        else:
            chains.append(i)
            count += 1
        
    return count
```

```python
count = count_chains(data['cadenaComercial'])
print(count)
```
